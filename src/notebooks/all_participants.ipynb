```python
import pandas as pd
from metabolics_slips_trips.loader import *
from metabolics_slips_trips.helper import *
from metabolics_slips_trips.plots import *
import matplotlib.pyplot as plt
```

```python
data = "../../data/"
```

```python
ds = DatasetVo2(data)
```

```python
ds.create_index()
```

# S1

```python
s1 = ds.get_subset(id=['S1'])
```

### time series data

```python
plt.figure(figsize = (8*1.5,3*6))
for s,i in enumerate(s1):
    plt.subplot(6,2,s*2+1)
    data = i.vo2_data.VO2
    plt.plot(data, label=i.index.magnitude[0])
    plt.vlines(i.impulse_times, data.min(), data.max(), color='red')
    plt.ylabel('vo2 [L/min]')
    plt.legend()

    plt.subplot(6,2,s*2+2)
    data = i.vo2_data.HR
    plt.plot(data, label=i.index.magnitude[0])
    plt.vlines(i.impulse_times, data.min(), data.max(), color='red')
    plt.ylabel('HR [bpm]')
    plt.legend()
plt.savefig('../../results/S1.png')
```

### all hr and vo2

```python
plt.figure()
for i in s1:
    t = i.hs_after_perturbation_times
    hr = i.vo2_data.HR
    res = all_responses(hr, t)
    mean_res = average_response(res)
    plt.plot(mean_res, label=i.index.magnitude[0])
plt.legend()
```

```python
plt.figure()
for i in s1:
    t = i.hs_after_perturbation_times
    vo2 = i.vo2_data.VO2
    res = all_responses(vo2, t)
    mean_res = average_response(res)
    plt.plot(mean_res, label=i.index.magnitude[0])
plt.legend()
    
```

### single example

```python
trial = s1[0]
```

```python
trial.index
```

```python
t = trial.hs_after_perturbation_times
hr = trial.vo2_data.HR
vo2 = trial.vo2_data.VO2
```

```python
rr = trial.vo2_data.VE
rr_response = all_responses(rr, t)
```

```python
hr_response = all_responses(hr, t)
vo2_response = all_responses(vo2, t)
```

```python
plt.figure()
x1 = plt.subplot(2,1,1)

subplot_response(x1, hr_response)
x1 = plt.subplot(2,1,2)
subplot_response(x1, vo2_response)
```

```python
plt.figure()
x1 = plt.subplot(2,1,1)
subplot_response(x1, rr_response)
```
